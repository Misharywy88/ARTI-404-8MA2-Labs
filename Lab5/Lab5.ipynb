## *Spatial Filtering – Smoothing and Sharpening* 🖼️✨








### Importing the Libraries  📚

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Task 1 📦

```python

def task1_box_filter():
    """
    Convolve an image with a 7x7 box filter, and output the 
    original and smoothed image side by side.
    """
    print("\n" + "=" * 60)
    print("📦 TASK 1: 7x7 Box Filter")
    print("=" * 60)
    
    # Read the image
    img = cv2.imread('Images/cat.jpg')
    
    # Check if image loaded
    if img is None:
        print("❌ Error: Could not load 'Images/cat.jpg'")
        return
    
    # Convert BGR to RGB
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # Create 7x7 box filter (averaging filter)
    kernel_size = 7
    box_kernel = np.ones((kernel_size, kernel_size), np.float32) / (kernel_size ** 2)
    
    # Apply the box filter
    smoothed = cv2.filter2D(img_rgb, -1, box_kernel)
    
    # Display results side by side
    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 2, 1)
    plt.imshow(img_rgb)
    plt.title('Original Image', fontsize=14, fontweight='bold')
    plt.axis('off')
    
    plt.subplot(1, 2, 2)
    plt.imshow(smoothed)
    plt.title('Smoothed Image (7x7 Box Filter)', fontsize=14, fontweight='bold')
    plt.axis('off')
    
    plt.suptitle('Task 1: Box Filter Smoothing', fontsize=16, fontweight='bold', color='navy')
    plt.tight_layout()
    plt.savefig('task1_box_filter.png', dpi=150, bbox_inches='tight')
    plt.show()
    
    print("✅ Task 1 completed!")
    print("💾 Saved: task1_box_filter.png")
```

### Task 2 🌀

```python
# ════════════════════════════════════════════════════════════════════════════
# TASK #2: Apply 5x5 and 21x21 Gaussian filters 
# ════════════════════════════════════════════════════════════════════════════
def task2_gaussian_filter():
    """
    Apply a 5x5 Gaussian filter to the image, then a 21x21 Gaussian filter,
    and show the three images side by side.
    """
    print("\n" + "=" * 60)
    print("🌀 TASK 2: Gaussian Filters (5x5 and 21x21)")
    print("=" * 60)
    
    # Read the image
    img = cv2.imread('Images/cat.jpg')
    
    # Check if image loaded
    if img is None:
        print("❌ Error: Could not load 'Images/cat.jpg'")
        return
    
    # Convert BGR to RGB
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # Apply 5x5 Gaussian filter
    gaussian_5x5 = cv2.GaussianBlur(img_rgb, (5, 5), 0)
    
    # Apply 21x21 Gaussian filter
    gaussian_21x21 = cv2.GaussianBlur(img_rgb, (21, 21), 0)
    
    # Display three images side by side
    plt.figure(figsize=(18, 6))
    
    plt.subplot(1, 3, 1)
    plt.imshow(img_rgb)
    plt.title('Original Image', fontsize=14, fontweight='bold')
    plt.axis('off')
    
    plt.subplot(1, 3, 2)
    plt.imshow(gaussian_5x5)
    plt.title('5x5 Gaussian Filter', fontsize=14, fontweight='bold')
    plt.axis('off')
    
    plt.subplot(1, 3, 3)
    plt.imshow(gaussian_21x21)
    plt.title('21x21 Gaussian Filter', fontsize=14, fontweight='bold')
    plt.axis('off')
    
    plt.suptitle('Task 2: Gaussian Filter Comparison', fontsize=16, fontweight='bold', color='navy')
    plt.tight_layout()
    plt.savefig('task2_gaussian_filter.png', dpi=150, bbox_inches='tight')
    plt.show()
    
    print("✅ Task 2 completed!")
    print("💾 Saved: task2_gaussian_filter.png")
```

### Task 3  🔪

```python
# ════════════════════════════════════════════════════════════════════════════
# TASK #3: Laplacian Sharpening (3x3)
# ════════════════════════════════════════════════════════════════════════════
def task3_laplacian_sharpening():
    """
    Apply a 3x3 Laplacian filter to sharpen the image.
    
    Steps:
    1. Read the image
    2. Convert image to float32
    3. Use the cv2.Laplacian() function
    4. Sharpen the image using: np.clip(image_float - laplacian, 0, 1)
    5. Plot the original, Laplacian result, and sharpened image
    """
    print("\n" + "=" * 60)
    print("🔪 TASK 3: Laplacian Sharpening (3x3)")
    print("=" * 60)
    
    # Step 1: Read the image
    print("   Step 1: Reading the image...")
    img = cv2.imread('Images/cat.jpg')
    
    # Check if image loaded
    if img is None:
        print("❌ Error: Could not load 'Images/cat.jpg'")
        return
    
    # Convert BGR to RGB
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # Step 2: Convert image to float32
    print("   Step 2: Converting image to float32...")
    image_float = img_rgb.astype(np.float32) / 255.0
    
    # Step 3: Use the cv2.Laplacian() function
    print("   Step 3: Applying cv2.Laplacian() function...")
    laplacian = cv2.Laplacian(image_float, cv2.CV_32F, ksize=3)
    
    # Step 4: Sharpen the image using: np.clip(image_float - laplacian, 0, 1)
    print("   Step 4: Sharpening image using np.clip(image_float - laplacian, 0, 1)...")
    sharpened = np.clip(image_float - laplacian, 0, 1)
    
    # Normalize Laplacian for visualization
    laplacian_display = cv2.normalize(laplacian, None, 0, 1, cv2.NORM_MINMAX)
    
    # Step 5: Plot the original, Laplacian result, and sharpened image
    print("   Step 5: Plotting results...")
    plt.figure(figsize=(18, 6))
    
    plt.subplot(1, 3, 1)
    plt.imshow(image_float)
    plt.title('Original Image', fontsize=14, fontweight='bold')
    plt.axis('off')
    
    plt.subplot(1, 3, 2)
    plt.imshow(laplacian_display)
    plt.title('Laplacian Filter Result', fontsize=14, fontweight='bold')
    plt.axis('off')
    
    plt.subplot(1, 3, 3)
    plt.imshow(sharpened)
    plt.title('Sharpened Image', fontsize=14, fontweight='bold')
    plt.axis('off')
    
    plt.suptitle('Task 3: Laplacian Sharpening', fontsize=16, fontweight='bold', color='navy')
    plt.tight_layout()
    plt.savefig('task3_laplacian_sharpening.png', dpi=150, bbox_inches='tight')
    plt.show()
    
    print("✅ Task 3 completed!")
    print("💾 Saved: task3_laplacian_sharpening.png")
```
